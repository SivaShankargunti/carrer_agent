```python
from langgraph.graph import MessageGraph,END
from langchain_core.messages import HumanMessage,AIMessage
from langchain_community.tools import TavilySearchResults
```

```python
search= TavilySearchResults(tavily_api_key=SEARCH_API_KEY)
def planner_node(state):
  user_input= state[-1].content
  prompt=("you are a smart planner. Given the user's question, decide if you need to perform an online search. If a search is needed, output ONLY the exact search query. If no search is needed, output ONLY 'NO_SEARCH'. Your output should be one or the other, with no additional text.") # Further refined prompt
  response= model.invoke([
      HumanMessage(content=prompt),
      HumanMessage(content=user_input)
  ]
  )
  return state + [AIMessage(content=response.content)]
def search_node(state):
  query=state[-1].content
  search_result=search.invoke(query)
  return state + [AIMessage(content=str(search_result))]


def responder_node(state):
  prompt=("yor are a helpful responser. use the following search result to the answer the users query")
  user_question=state[0].content
  search_result=state[-1].content
  response= model.invoke([
      HumanMessage(content=prompt),
      HumanMessage(content=f"user question: {user_question} and search result: {search_result}"),

  ]
  )
  return state + [AIMessage(content=response.content)]




graph=MessageGraph()
graph.add_node("planner",planner_node)
graph.add_node("search",search_node)
graph.add_node("responder",responder_node)
graph.set_entry_point("planner")

graph.add_edge("planner","search")
graph.add_edge("search","responder")
graph.add_edge("responder",END)


agent=graph.compile()
agent_response=agent.invoke([HumanMessage(content="who is the siva shankar gunti")])
print(agent_response)
```

```python
from langgraph.graph import MessageGraph, END
from langchain_core.messages import HumanMessage, AIMessage
from langchain_core.prompts import ChatPromptTemplate

# Prompt for writing a LinkedIn post
write_prompt = ChatPromptTemplate.from_template(
    "write a linkedin post about this topic: {topic}"
)

# Chain
write_chain = write_prompt | model

# Graph
graph = MessageGraph()

# Writer node
def writer_node(state):
    print("\n--- Writer Node ---")
    topic = state[-1].content
    post = write_chain.invoke({"topic": topic})
    print(f"Writer generated post:\n{post.content}\n")
    return state + [AIMessage(content=post.content)]

# Reviewer node
def reviewer_node(state):
    print("\n--- Reviewer Node ---")
    post = state[-1].content
    print(f"Post to review:\n{post}\n")
    review_input = input("Enter your review: ")  # <-- Keyboard input
    return state + [HumanMessage(content=review_input)]

# Add nodes
graph.add_node("writer", writer_node)
graph.add_node("reviewer", reviewer_node)

# Set entry point
graph.set_entry_point("writer")

# Conditional edge check
def condition_check(state):
    if len(state) >= 5:  # adjust condition as needed
        print(">>> Ending process.")
        return END
    else:
        print(">>> Sending to reviewer.")
        return "reviewer"

graph.add_conditional_edges("writer", condition_check)
graph.add_edge("reviewer", "writer")

# Compile app
app = graph.compile()

# Run
response = app.invoke([HumanMessage(content="get the detail movie review about kuberaa")])
for message in response:
    print(message.type, ":", message.content)
```

# State graph example

```python
from langgraph.graph import StateGraph,END
from typing import TypedDict
class mystate(TypedDict):
  count:int
graph = StateGraph(mystate)
def increament_node(state:mystate):
  print("inside increament_node,current state=",state)
  state["count"]=state["count"]+1
  return state

graph.add_node("increament",increament_node)
graph.set_entry_point("increament")
def iscountreachedlimit(state:mystate):
  if state["count"]>=8:
    return END
  else:
    return "increament"
graph.add_conditional_edges("increament",iscountreachedlimit)
app=graph.compile()
response=app.invoke({"count":0})
print(response)
```

#chat bot with memory using checkpoint


```python
from langgraph.graph import StateGraph, END
from langchain_core.messages import HumanMessage, AIMessage
from typing import TypedDict, Annotated
from langgraph.checkpoint.memory import MemorySaver
import re # Import the re module

# --- Define State ---
class MessagesState(TypedDict):
    messages: Annotated[list, "messages"]
    memory: dict   # <-- to store facts like "name"

graph = StateGraph(MessagesState)
memory = MemorySaver()

# --- Helper: extract name ---
def extract_name(text: str):
    match = re.search(r"my name is (\w+)", text, re.IGNORECASE)
    if match:
        return match.group(1)
    return None

# --- Chat node ---
def chat_node(state: MessagesState):
    print("inside chat_node, current state =", state)

    user_message = state["messages"][-1].content
    name = extract_name(user_message)

    # Initialize memory if it doesn't exist
    if "memory" not in state or state["memory"] is None:
        state["memory"] = {}

    if name:  # user tells their name
        state["memory"]["name"] = name
        reply = f"Nice to meet you, {name}! I'll remember your name."
    elif "what is my name" in user_message.lower():
        if "name" in state["memory"]:
            reply = f"Your name is {state['memory']['name']}."
        else:
            reply = "I don’t know your name yet. Please tell me."
    else:
        if "name" in state["memory"]:
            reply = f"Hello {state['memory']['name']}, how can I help you?"
        else:
            reply = "Hello! You can tell me your name and I'll remember it."

    response = AIMessage(content=reply)
    return {
        "messages": state["messages"] + [response],
        "memory": state["memory"]
    }

# --- Build graph ---
graph.add_node("chat", chat_node)
graph.set_entry_point("chat")
graph.add_edge("chat", END)

agent = graph.compile(checkpointer=memory)

Configuration = {"configurable": {"thread_id": "1"}}

# --- Example run ---
ans1 = agent.invoke(
    {"messages": [HumanMessage(content="my name is Siva")], "memory": {}}, # Pass initial memory state
    config=Configuration
)
print(ans1)

ans2 = agent.invoke(
    {"messages": [HumanMessage(content="what is my name?")], "memory": ans1["memory"]}, # Pass the updated memory state
    config=Configuration
)
print(ans2)

ans3 = agent.invoke(
    {"messages": [HumanMessage(content="hi again")], "memory": ans2["memory"]}, # Pass the updated memory state
    config=Configuration
)
print(ans3)
```
